```python
## Computing Volatility

# Load the required modules and packages
import numpy as np
import pandas as pd
import pandas_datareader.data as web
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
frame = pd.read_csv('Plotting_green_investment.csv')
```

```python
frame = frame.set_index('yearly_port')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
 
# Make a data frame
frame = frame.rename(columns = {'yearly_ret_port_opt':'Optimal predicted weight portfolio return',
                       'yearly_ret_port_equal':'Equal weight portfolio return','yearly_ret_s&p':'S&P500 return'
                                ,'min_port_opt':'Minimum optimal predicted portfolio return'
                               ,'max_port_opt':'Maximum optimal predicted portfolio return'})

df= frame[['Minimum optimal predicted portfolio return','Maximum optimal predicted portfolio return'
           ,'S&P500 return','Equal weight portfolio return','Optimal predicted weight portfolio return']]

plt.style.use('seaborn-darkgrid')
my_dpi=96
plt.figure(figsize=(480/my_dpi, 480/my_dpi), dpi=my_dpi)

ax = plt.bar(df.index, df['Equal weight portfolio return']*100, color='red', linewidth=2, alpha=0.4)
ax = plt.bar(df.index, df['S&P500 return']*100, color='orange', linewidth=2, alpha=0.4)
 
# Now re do the interesting curve, but biger with distinct color
ax = plt.bar(df.index, df['Optimal predicted weight portfolio return']*100, color='green', linewidth=3, alpha=0.7)

ax = plt.plot(df.index, df['Minimum optimal predicted portfolio return']*100, color='green', linewidth=1, alpha=0.3,marker='')
ax = plt.plot(df.index, df['Maximum optimal predicted portfolio return']*100, color='green', linewidth=1, alpha=0.3,marker='')


plt.title("Return plot of Optimal predicted weight portfolio return", loc='center', fontsize=12, weight = 'bold') #, color='orange')
plt.xlabel("Year", fontsize = 12, weight = 'bold')
plt.ylabel("%Return", fontsize = 12, weight = 'bold')
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.legend(df.columns, fontsize = 8)
plt.show()

 
```

```python
frame = frame.rename(columns = {'sd_port_opt':'Optimal weight predicted portfolio volatility'
                                , 'sd_port_equal':'Equal weight predicted portfolio volatility'
                                , 'sd_s&p':'S&P500 volatility'})

df= frame[['S&P500 volatility','Equal weight predicted portfolio volatility'
           ,'Optimal weight predicted portfolio volatility']]

plt.style.use('seaborn-darkgrid')
my_dpi=96
plt.figure(figsize=(480/my_dpi, 200/my_dpi), dpi=my_dpi)

ax = plt.plot(df.index, df['S&P500 volatility']*100, color='red', linewidth=2, alpha=0.4)
ax = plt.plot(df.index, df['Equal weight predicted portfolio volatility']*100, color='orange', linewidth=2, alpha=0.4)
 
# Now re do the interesting curve, but biger with distinct color
ax = plt.plot(df.index, df['Optimal weight predicted portfolio volatility']*100, color='green', linewidth=3, alpha=0.7)

plt.title("Return plot of Optimal predicted weight portfolio return", loc='center', fontsize=12, weight = 'bold') #, color='orange')
plt.xlabel("Year", fontsize = 12, weight = 'bold')
plt.ylabel("%Volatility", fontsize = 12, weight = 'bold')
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
#plt.legend(df.columns, fontsize = 9)
plt.show()

 
```
